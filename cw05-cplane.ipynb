## Abstract Base Classes
The Python Abstract Base Class (ABC) is python's mechanism for implementing what is known as an Inteface in other OOP languages.  ABC can be used to set up a template, whereby other classes derived from the template will have a set of methods that the derived class should implement in order to complete the implementation of the class.  If an interface method is overlooked (not implemented), then it is likely that an error will be generated early in the project development, allowing faster and easier correction of the problem than if the problem is not detected until much later in the project.

## Complex Plane ABC Example
Our ComplexPlane class is derived from the Abstract Base Class AbsComplexPlane.  AbsComplexPlane sets the template interface to require the methods refresh(), zoom(), and set_f().  In addition, there is need for a creator function to set the initial parameters of the complex plane.
<br>In our implementation of ComplexPlane, the creator, `__init__()`, takes 4 arguments to set the 4 corners of the Real and Imaginary plane.  The creator generates a matrix to represent the plane and populates the values of the 2-dimensional plane with the coordinates of each point in the plane.
<br>set_f() takes a method as its argument and will call that method with the point coordinates in the plane to perform the transformation of the points through whatever function that the method is programmed to do.  The default transformation function for a newly created ComplexPlane object is simply an identity transform.
<br>refresh() uses list comprehension to sequence through alll the points in the ComplexPlane and recalculates the values at each point using the current transformation function set using set_f().
<br>zoom() takes 4 arguments to allow the caller to effectively pan and zoom over the imaginary ComplexPlane.  A new plane area is set up and the point values recalculated through a call to refresh().

```python
# set the min/max size of the example complex plane
xmin = -5.
xmax = 5.
ymin = -5.
ymax = 5.
```

```python
# create the plane
thePlane = cpl.ComplexPlane(xmin,xmax,ymin,ymax)
```
